```python
import pandas as pd

# Load the dataset
auto_mpg = pd.read_csv('auto-mpg.csv')

# Display the first few rows
auto_mpg.head()
```

```python
# Check for missing values
missing_values = auto_mpg.isnull().sum()
missing_values
```

```python
# Skewness
skewness = auto_mpg.skew()
print("Skewness:\n", skewness)

# Kurtosis
kurtosis = auto_mpg.kurtosis()
print("\nKurtosis:\n", kurtosis)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
corr_matrix = auto_mpg.corr()

# Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Example: MPG vs. Weight
sns.scatterplot(x='weight', y='mpg', data=auto_mpg)
plt.title('MPG vs. Weight')
plt.show()

# You can add more scatter plots for different pairs of variables
```
