```python
import pandas as pd

# Define column names
column_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
                'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']

# Load dataset
diabetes_data = pd.read_csv('/content/pima-indians-diabetes.data.csv', names=column_names)

# Display the first few rows
diabetes_data.head()
```

```python
# Basic info about the dataset
diabetes_data.info()

# Check for missing values
diabetes_data.isnull().sum()

# Statistical summary
diabetes_data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for Glucose
diabetes_data['Glucose'].hist(bins=15)
plt.title('Glucose Distribution')
plt.xlabel('Glucose Level')
plt.ylabel('Frequency')
plt.show()

# Boxplot for BMI
sns.boxplot(y=diabetes_data['BMI'])
plt.title('Boxplot of BMI')
plt.show()
```

```python
# Correlation matrix
corr = diabetes_data.corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pair plot for a subset of variables
sns.pairplot(diabetes_data[['Glucose', 'BloodPressure', 'BMI', 'Age', 'Outcome']], hue='Outcome')
plt.show()
```
